```python
import numpy as np
import os
import json
from PIL import Image
from wordcloud import WordCloud, ImageColorGenerator
import matplotlib.pyplot as plt
%matplotlib inline

s = json.load(open(os.getenv('PUI')+'plotstyle.json'))
plt.rcParams.update(s)
```

```python
np.random.seed(2015)
path = os.getenv('WORDCLOUD')+'/JuliaLaneDataHygeneTalk.txt'
text = open(path).read()
#create mask (outline of the wordcloud)
path_mask = np.array(Image.open('titan.jpg'))
#create the color of the wordcloud based on a image
path_color = np.array(Image.open('titan.jpg'))
wdColors = ImageColorGenerator(path_color)
# cross_mask = np.array(Image.open('cross.jpg'))
wordcloud = WordCloud(ranks_only=True, scale = 4, mask = path_mask,
                      min_font_size=10, max_font_size=200, 
                      font_step = 10, max_words = None,
                      background_color = 'black').generate(text)
plt.figure(figsize=(15,15))
plt.imshow(wordcloud.recolor(color_func=wdColors))
plt.imshow(wordcloud, cmap=plt.cm.cool)
plt.axis('off')
```
